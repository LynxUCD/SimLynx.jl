# SimLynx.jl Tally and Accumulate Example

This example demonstrates the differences between tallied and accumulated variables for automatic data collection. 

```julia
using SimLynx
SimLynx.greet()
```

## Data Collection

The global variables ``tallied`` and ``accumulated`` are used to tally and accumulate the sample data items. The variables are initialized in the main program.

```julia
tallied = nothing
accumulated = nothing
```

## Active Elements

The ``test_process(value_durations)`` process iterates over the ``value_durations``, sets the ``tallied`` and ``accumulated`` variables to the specified value, and then works for the specified duration.

```julia
@process test_process(value_durations) begin
    for (value, duration) in value_durations
        tallied.value = value
        accumulated.value = value
        work(duration)
    end
end
```

## Main Program

The main program, ``main(value_durations)``, creates a new simulation, creates the ``tallied`` and ``accumulated``, schedules the ``test_process`` to start at time 0.0, and starts the simulation. The ``tallied`` variable is set to tally its data and to maintain a history of its values. The ``accumulated`` variable is set to accumulate (the default) its data and to maintain a history of its values. Both variables will also maintain running statistics on its values, which is also the default.

After the simulation ends, the values of ``n``, ``sum``, and ``mean`` are printed for both the ``tallied`` and ``accumulated``, and their histories are plotted.

```julia
function main(value_durations)
    println("--- Test Tally and Accumulate ---")
    @simulation begin
        # Create tallied and accumulated variables
        global tallied = Variable{Int64}(data=:tally, history=true)
        global accumulated = Variable{Int64}(0, history=true)
        # Schedule the test process and start the simulation
        @schedule at 0.0 test_process(value_durations)
        start_simulation()
        # Print and plot the tallied results
        println("--- Tally ---")
        print_stats(tallied, title="Tallied Statistics")
        plot_history(tallied, title="Tallied History")
        # Print and plot the accumulated results
        println("--- Accumulate ---")
        print_stats(accumulated, title="Accumulated Statistics")
        plot_history(accumulated, title="Accumulated History")
    end
end
```

## Output

This example demonstrates tally and accumulated for four (4) value/duration pairs. These are:
* value of one (1) for 2.0 time units
* value of two (2) for 1.0 time units
* value of three (3) for 2.0 time units
* value of four (4) for 3.0 time units

```julia
main([(1, 2.0), (2, 1.0), (3, 2.0), (4, 3.0)])
```

The tallied results are independant of time. There are four (4) values (1, 2, 3, and 4) so:
* ``n = 4``
* ``sum = 1 + 2 + 3 + 4 = 10``
* ``mean = sum / n = 10 / 4 = 2.5``

The accumulated results are time dependant. That is, the values are accumulated (i.e., summed or integrated) over time. There are four (4) values (1, 2, 3, and 4) with durations of 2.0, 1.0, 2.0, and 3.0 so:
* ``n = 8.0`` (the total duration)
* ``sum = 1 * 2.0 + 2 * 1.0 + 3 * 2.0 + 4 * 3.0 = 22.0`` (the area under the curve)
* ``mean = sum / n = 22.0 / 8.0 = 2.75``
